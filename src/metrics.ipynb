```python
import pandas as pd
import matplotlib.pyplot as plt
from collections import Counter
```

```python
donors = pd.read_csv('../data/donors.csv')
```

```python
donors
```

```python
donors.blood_type.unique()
```

```python


# Convert the donors list to a pandas DataFrame
df = pd.DataFrame(donors)

# Ensure the columns are of appropriate types
df['age_at_donation'] = df['age_at_donation'].astype(int)
df['sex'] = df['sex'].astype(str)
df['ethnicity'] = df['ethnicity'].astype(str)
df['blood_type'] = df['blood_type'].astype(str)

# Plot age distribution
plt.figure(figsize=(10, 6))
df['age_at_donation'].hist(bins=range(17, 82), edgecolor='k', alpha=0.7)
plt.title('Age Distribution of Donors')
plt.xlabel('Age at Donation')
plt.ylabel('Number of Donors')
plt.show()

# Plot sex distribution
plt.figure(figsize=(10, 6))
df['sex'].value_counts().plot(kind='bar', edgecolor='k', alpha=0.7)
plt.title('Sex Distribution of Donors')
plt.xlabel('Sex')
plt.ylabel('Number of Donors')
plt.show()

# Plot ethnicity distribution
plt.figure(figsize=(10, 6))
df['ethnicity'].value_counts().plot(kind='bar', edgecolor='k', alpha=0.7)
plt.title('Ethnicity Distribution of Donors')
plt.xlabel('Ethnicity')
plt.ylabel('Number of Donors')
plt.show()

# Plot blood type distribution
plt.figure(figsize=(10, 6))
df['blood_type'].value_counts().plot(kind='bar', edgecolor='k', alpha=0.7)
plt.title('Blood Type Distribution of Donors')
plt.xlabel('Blood Type')
plt.ylabel('Number of Donors')
plt.show()

```
